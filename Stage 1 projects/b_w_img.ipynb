# convert an image to black and white

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

# use imread.grayscale to load the img into grayscale first instance
# different from COLOR_BGR2GRAY, which conv img scale AFTER it loads in

img = cv.imread('Images/ER.jpg', cv.IMREAD_GRAYSCALE)

_, thresholded = cv.threshold(img, 127, 255, cv.THRESH_BINARY) # below 127: black, above 127: white

plt.imshow(thresholded, cmap='gray') # no need to swap bgr to rbg, since we're dealing with just black and white
```
